```python
## 1. Load json file to Panda's dataframe to do some Calculate metrics for some dimensions
  
import pandas as pd
# Load into panda's dataframe for metrices - Impressions(df_Im)
df_Im = pd.read_json (r'impressions.json')
print (df_Im)

# Load into panda's dataframe for metrices - Clicks(Ck_Im)
df_Ck = pd.read_json (r'clicks.json')
print (df_Ck)
```

```python
## 2.  Calculate metrics for some dimensions
# Count of impressions
Count_Im = len(df_Im)
print ("Impression count is ", Count_Im)
# 2047

#Count of clicks
Count_Ck= len(df_Ck)
print ("Click count is ", Count_Ck)
# 95

#Sum of revenue
Sum_Re = df_Ck['revenue'].sum()
print ("Sum of revenue is ", Sum_Re)
```

```python
## 2.  Calculate metrics for some dimensions
df2= dfinal.groupby(['app_id', 'country_code']).apply(lambda s: pd.Series({
                "Impress_count": s["id"].count(),
                "Clicks_count" : s["revenue"].value_counts().count(),
                "Revenue_sum"  : s["revenue"].sum()
}))

print(df2)
```

```python
## Quesstion 3 importin json file into spark pre-importing and ENV setup
from pyspark import SparkContext
from pyspark.sql import SparkSession 
from pyspark.sql import SQLContext
import os
os.environ['PYSPARK_SUBMIT_ARGS'] = "spark://ML-PC:7077"

sc = SparkContext.getOrCreate()
spark = SparkSession(sc)
sqlc = SQLContext(sc)
```

```python
#Import impressions.json file to Spark
df_imp = spark.read.option("multiline","true").json("impressions.json")
#print(df_imp.schema)
df_imp.show()
#Import clicks.json file to Spark
df_clk = spark.read.option("multiline","true").json("clicks.json")
df_clk.show()
```

```python
# SQL Joining two tables with column - id 
df_imp.createOrReplaceTempView("IMP")
df_clk.createOrReplaceTempView("CLK")

imp_clk=spark.sql("select * from IMP i, CLK c where i.id == c.impression_id")
imp_clk.show()
```

```python
# Adding column to Joined table - imp_clk with Performance
imp_clk.registerTempTable("imp_clk")
#imp_clk_perf=sqlc.sql("select app_id,country_code,count(id) as Impress_count,count(distinct revenue) as Clicks_count,sum(revenue) as Revenue_sum, sum(revenue)/count(id) as Performance  from imp_clk group by app_id,country_code")
imp_clk_perf=sqlc.sql("select app_id,country_code,advertiser_id,count(id) as Impress_count,count(distinct revenue) as Clicks_count,sum(revenue) as Revenue_sum, sum(revenue)/count(id) as Performance  from imp_clk group by app_id,country_code,advertiser_id")
imp_clk_perf.show(10)
```
